```python
### HYPERUNIFORMITY ###

import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image
%matplotlib widget
```

```python
def func(x):
    return 0.457648/x
```

```python
# Delta = 0 for L = 10, 100, 200

Ls = [10,100,200]
d = 0

plt.figure()
Rs = []
scvars = []
for L in Ls:
    with open("scaledvariance_L"+str(L)+"_d"+str(d)+".dat","r") as f:
        lines = f.readlines()
        R = []
        scvar = []
        for line in lines[1:]:
            splitline = line.split()
            R.append(float(splitline[0]))
            scvar.append(float(splitline[1]))
    Rs.append(R)
    scvars.append(scvar)
    plt.plot(R,scvar,'.-',markersize = 3, linewidth = 1, alpha = 0.5, label="L = "+str(L))
xpoints = np.arange(0.8,100,0.1)
plt.plot(xpoints, func(xpoints), label = "$\sigma^2(R)/R = 0.457648/R$")
plt.legend()
plt.grid(True, which="both")
plt.xlabel("$R$")
plt.ylabel("$\sigma^2(R)/R^2$")
plt.xscale('log')
plt.yscale('log')
plt.savefig('ScaledVarianceEx1.pdf')
plt.show()
```

$\textit{Question:}$  why does $\sigma^2(R)/R^2$ goes to a constant at short distance?

If we consider short $R$,$i.e. R<a/2$, we realize that a random point taken in the lattice can either have $1$ or $0$ neighbours. This situation represent a Bernoulli process, in which the probability of success is the probability of having one neighbour (i.e. to extract a random point which has a lattice point at distance < $R$). Then we have $P(success)= \frac{\pi R^2}{a^2} = \pi R^2$. The variance is then given by $P(success)*(1-P(success))$ which is equal to $R^2\pi(1-\pi R^2)$. What we plotted above is the variance divided by $R^2$, so at the and we have $\sigma^2/R^2 = \pi(1-\pi R^2)$ for $R<=a/2$. To verify our prediction we can fit the above results for $L=10$ with a parabola of the form $ax^2+b$, obtaining as expected $a\approx -9.45$ and $b\approx 3.141$.

$\textit{Question:}$ What are the reasons of the oscillations of the curve?

The oscillations are due to the breaking of translational invariance in the lattice. Since the different point are separated by a well definite length scale (the lattice constant), $R$ must increase of a certain amount before the maximum number of neighbours increases of one unit, as it was seen in the example with $R<a/2$. This would not be the case if $a\to 0$ or if $R\to\infty$. Also, changing the lattice parameter causes a variation in the oscillation frequency. \\DA VERIFICARE

$\textit{Question:}$Why do the amplitudes of the oscillations decrease at long distance?

As stated before, for $R\to\infty$ the system approaches a continuum limit, hence it takes a smaller increment of $R$ to increase of one unit the maximum number of accessible neighbours for one point.

```python
# Delta = 0, 1, 2, 3, 4, 5 for L = 100

L = 100
ds = [0, 1, 2, 3, 4, 5]

plt.figure()
Rs = []
scvars = []
for d in ds:
    with open("scaledvariance_L"+str(L)+"_d"+str(d)+".dat","r") as f:
        lines = f.readlines()
        R = []
        scvar = []
        for line in lines[1:]:
            splitline = line.split()
            R.append(float(splitline[0]))
            scvar.append(float(splitline[1]))
    Rs.append(R)
    scvars.append(scvar)
    plt.plot(R,scvar,'.-',markersize = 3, linewidth = 1, alpha = 0.5, label="$\Delta$ = "+str(d))
xpoints = np.arange(0.8,L/2,0.1)
#plt.plot(xpoints, func(xpoints), label = "$\sigma^2(R)/R = 0.457648/R$")
plt.legend()
plt.grid(True, which="both")
plt.xlabel("$R$")
plt.ylabel("$\sigma^2(R)/R^2$")
plt.xscale('log')
plt.yscale('log')
plt.savefig('ScaledVarianceEx2.pdf')
plt.show()
```

$\textit{Question:}$ What do you observe?

For $\Delta>=1$ the oscillations disappear since the random nature of the new lattice (thanks to the shuffle) allows for a continuum growth of the maximum number of accessible neighbours within a certain $R$, thus mimicking a continuus system and not a discrete one as for $\Delta = 0$.
